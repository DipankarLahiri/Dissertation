```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

base_url = "https://sanatanprabhat.org/english/category/post-type/news/international/page/{}"
data = []

headers = {
    "User-Agent": "Mozilla/5.0"
}

for page_num in range(23, 51):  # inclusive of page 50
    url = base_url.format(page_num)
    print(f"Scraping page {page_num}: {url}")

    response = requests.get(url, headers=headers)
    if response.status_code != 200:
        print(f"Failed to load page {page_num}")
        continue

    soup = BeautifulSoup(response.content, "html.parser")
    articles = soup.find_all("h2", class_="entry-title")

    for article in articles:
        a_tag = article.find("a")
        if not a_tag:
            continue

        title = a_tag.text.strip()
        link = a_tag["href"]

        # Now get the date from the associated <time> tag
        time_tag = article.find_next("time", class_="entry-date")
        date = time_tag.text.strip() if time_tag else "N/A"

        data.append({
            "headline": title,
            "link": link,
            "date": date
        })

    time.sleep(1)  # Be polite to the server

# Save to CSV
df = pd.DataFrame(data)
df.to_csv("sanatanprabhat_headlines_p23to50.csv", index=False)
```
